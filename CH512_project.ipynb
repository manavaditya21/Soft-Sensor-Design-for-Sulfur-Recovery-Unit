```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
x = x = np.array(df1['IN1']).reshape(-1, 1)  # Reshaping to make it a 2D array
y = df2['Out1']

# Degree of the polynomial
degree = 5

# Transforming input features to include polynomial terms up to degree
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)

# Fit polynomial regression model
model = LinearRegression()
model.fit(x_poly, y)

# Predictions
y_pred = model.predict(x_poly)

# Sort x and y_pred for plotting purposes
sorted_zip = sorted(zip(x, y_pred))
x_sorted, y_pred_sorted = zip(*sorted_zip)

# Plot original data
plt.scatter(x, y, label='Data')

# Plot polynomial regression curve
plt.plot(x_sorted, y_pred_sorted, color='red', label='Polynomial Regression (degree {})'.format(degree))

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Polynomial Regression')

# Add legend
plt.legend()

# Show plot
plt.show()
```
